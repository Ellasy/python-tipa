### 1.1 Задание

```python
import numpy as np
import math

class Trigonometry:
    def __init__(self) -> None:
        pass

    def calculate(self, command: str) -> float:
        split_cmd = command.strip().split(' ')
        return self.__inner_calculate__(split_cmd)
        

    def __inner_calculate__(self, split_cmd) -> float:
        reduced_str = ' '.join(split_cmd[1:]).strip().split()
        
        if split_cmd[0] == 'cos':
            print('cos')
            return np.cos(self.__inner_calculate__(reduced_str))
        elif split_cmd[0] == 'sin':
            print('sin')
            result = np.sin(self.__inner_calculate__(reduced_str))
            return result
        elif split_cmd[0] == 'tg':
            print('tg')
            return np.tan(self.__inner_calculate__(reduced_str))
        elif split_cmd[0] == 'arccos':
            print('arccos')
            return np.arccos(self.__inner_calculate__(reduced_str))
        elif split_cmd[0] == 'arcsin':
            print('arcsin')
            return np.arcsin(self.__inner_calculate__(reduced_str))
        elif split_cmd[0] == 'arctg':
            print('arctg')
            return np.arctan(self.__inner_calculate__(reduced_str))
        elif split_cmd[0] == 'rad':
            print('rad')
            return math.radians(self.__inner_calculate__(reduced_str))
        elif split_cmd[0] == 'deg':
            print('deg')
            return math.degrees(self.__inner_calculate__(reduced_str))
        else:
            return float(split_cmd[0])

    
```

### 1.2.2 Задание

```python
class Tree:
    def __init__(self, data):
        self.left = None 
        self.right = None 
        self.data = data
        
    def PrintTree(self): 
        print(self.data)

    def print_tree(self):
        
        print(f'height: {self.height()} data: {self.data}')
        if self.left is not None:
            self.left.print_tree()
        if self.right is not None:
           self.right.print_tree()
        
        #print(self.collect_data())

    def collect_data(self):
        layers_data = []

        if self.left is not None:
            layers_data.append(self.left.collect_data())
        if self.right is not None:
           layers_data.append(self.right.collect_data())

        layers_data.append({self.height(): self.data})

        return layers_data

    def height(self):
        if self.left is None:
            l_height = 0
        else:
            l_height = self.left.height() 

        if self.right is None:
            r_height = 0
        else:
            r_height = self.right.height()
        return max(l_height, r_height) + 1

    def append_element(self, tree):
        if tree.data < self.data:
            if self.left is None:
                self.left = tree
            else:
                self.left.append_element(tree)
        elif tree.data > self.data:
            if self.right is None:
                self.right = tree
            else:
                self.right.append_element(tree)
        else:
            pass
```

```python
tree = Tree(23)

for i in range(3, 39, 3):
    tree.append_element(Tree(i))

for i in range(1, 39, 3):
    tree.append_element(Tree(i))
    
tree.print_tree()
```

### 1.3 Задание

```python
from sklearn.model_selection import train_test_split

X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 2], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

X_train, X_test, y_train, y_test = train_test_split(
    X, 
    target, 
    test_size=0.2
    )

X_train, y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)

tree.plot_tree(classifier)
```

```python
y_pred = classifier.predict(X_test)
y_pred, y_test
```
